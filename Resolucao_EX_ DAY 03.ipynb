#Importando as informações para os exercícios

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('salarios_br(2).csv', sep = ',')
dataset.head(5)
```

#1. Plotando o gráfico de barras ('P5', 'living_state')

```python
dataset["('P5', 'living_state')"].value_counts().plot.barh(edgecolor = "black")
plt.show()
```

#2. Plotando o gráfico de barras ('P2', 'gender')

```python
dataset["('P2', 'gender')"].value_counts().plot.barh(color=['blue', 'pink'], edgecolor = "black")
plt.show()
```

#3. Plotando o gráfico de barras ('P22', 'most_used_proggraming_languages')

```python
dataset["('P22', 'most_used_proggraming_languages')"].value_counts().plot.barh(edgecolor = "black")
plt.show()
```

#4. Plotando o gráfico de barras ('P10', 'job_situation')

```python
dataset["('P10', 'job_situation')"].value_counts().plot.barh(edgecolor = "black")
plt.show()
```
